```python
% matplotlib notebook 

import numpy as np
import matplotlib.pyplot as plt
from pyuvdata import UVData
from matplotlib.colors import SymLogNorm
import os
import glob

plt.rc('font', family='serif')
plt.rc('xtick', labelsize='small')
plt.rc('ytick', labelsize='small')
```

```python
deci = 3

time_avg_baselines_nf = np.load('time_avg_baselines_nf.npy')
baseline_arr_nf = np.load('baseline_length_arr_nf.npy')

baseline_round_nf = np.unique(np.array(baseline_arr_nf)[:2070].round(deci))
baseline_round.sort()

rounded = np.array(baseline_arr_nf)[:2070].round(deci)

bases_nf = []

for i in baseline_round_nf:
    bases_nf.append(np.mean(time_avg_baselines_nf[np.where(rounded == i)],axis=0))

bases_nf = np.array(bases_nf)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(bases_nf.T[:512,],aspect='auto',norm=SymLogNorm(vmin=1e3,vmax=1e7,linthresh=.1))
plt.xticks(np.arange(0,baseline_round_nf.shape[0],20),baseline_round_nf[::20].round(1))
plt.xlabel('Baseline Length (m)')
plt.yticks(np.arange(0,baseline_round.shape[0],30),baseline_round[::30].round(1))
plt.ylabel('Delay (ns)')
plt.colorbar()
```

```python
deci = 3

time_avg_baselines = np.load('time_avg_baselines.npy')
baseline_arr = np.load('baseline_length_arr.npy')

baseline_round = np.unique(np.array(baseline_arr)[:2070].round(deci))
baseline_round.sort()

rounded = np.array(baseline_arr)[:2070].round(deci)

bases = []

for i in baseline_round:
    bases.append(np.mean(time_avg_baselines[np.where(rounded == i)],axis=0))

bases = np.array(bases)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(bases.T[:512,],aspect='auto',norm=SymLogNorm(vmin=1e3,vmax=1e7,linthresh=.1))
plt.xticks(np.arange(0,baseline_round.shape[0],20),baseline_round[::20].round(1))
plt.xlabel('Baseline Length (m)')
plt.yticks(np.arange(0,baseline_round.shape[0],30),baseline_round[::30].round(1))
plt.ylabel('Delay (ns)')
plt.colorbar()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(bases_nf.T[:512,],aspect='auto',norm=SymLogNorm(vmin=1e3,vmax=1e7,linthresh=.1))
plt.colorbar()
plt.xticks(np.arange(0,baseline_round_nf.shape[0],30),baseline_round_nf[::30].round(1))
plt.xlabel('Baseline Length (m)')
plt.title('Unfiltered Delay Wedge')
plt.yticks(np.arange(0,baseline_round.shape[0],30),baseline_round[::30].round(1))
plt.ylabel('Delay (ns)')

plt.subplot(1,2,2)
plt.imshow(bases.T[:512,],aspect='auto',norm=SymLogNorm(vmin=1e3,vmax=1e7,linthresh=.1))
plt.colorbar()
plt.xticks(np.arange(0,baseline_round.shape[0],30),baseline_round[::30].round(1))
plt.xlabel('Baseline Length (m)')
plt.yticks(np.arange(0,baseline_round.shape[0],30),baseline_round[::30].round(1))
plt.ylabel('Delay (ns)')
plt.title('Delay Filtered Delay Wedge')
```

```python
plt.figure(figsize=(10,5))
plt.imshow(bases_nf.T[:512,]-bases.T[:512,],aspect='auto',norm=SymLogNorm(vmin=1e3,vmax=1e7,linthresh=.1))
plt.title('Unfilter-Delay Filtered Difference Plot')
plt.xticks(np.arange(0,baseline_round_nf.shape[0],20),baseline_round_nf[::20].round(1))
plt.xlabel('Baseline Length (m)')
plt.yticks(np.arange(0,baseline_round.shape[0],30),baseline_round[::30].round(1))
plt.ylabel('Delay (ns)')
plt.colorbar()
```
